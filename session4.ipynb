```python
#Task 1: The "Safe Lookup" Dictionary Engine

# create dictionary

inventory = {
"sku_01": 50,
"sku_02": 150,
"sku_03": 75
}
```

```python
# OPERATIONS
# update()
inventory.update({"sku_04": 200})
```

```python
# Retrieve
sku_02_stock = inventory["sku_02"]
print("Stock of sku_02:", sku_02_stock)
```

```python
# get() method to search for "sku_05"
sku_05_stock = inventory.get("sku_05", 0)
print("Stock of sku_05:", sku_05_stock)
```

```python
# use key() and value() to print list of all items
print
(
"Product IDs:"
,
inventory.keys
())
print
(
"Stock counts:"
,
inventory.values
())
```

```python
# Convert dictionary to a list of tuples using .items()

inventory_items =list(inventory.items())
print("Inventory as list of tuples:",inventory_items)
```

```python

#Task 2: Deep Access in Nested JSON-like Structures

# Data setup
api_response = {
    "status": "success",
    "data": [
        {
            "id": 1,
            "info": {
                "email": "user1@work.com",
                "tags": ["admin", "dev"]
            }
        },
        {
            "id": 2,
            "info": {
                "email": "user2@work.com",
                "tags": ["guest"]
            }
        }
    ]
}

# Access the email of the first user
user_1_email = api_response["data"][0]["info"]["email"]
print("User 1 Email:", user_1_email)

# Access the second tag of the first user ("dev")
user_1_secondary_tag = api_response["data"][0]["info"]["tags"][1]
print("User 1 Secondary Tag:", user_1_secondary_tag)

# Retrieve the ID of the last user using negative indexing
last_user_id = api_response["data"][-1]["id"]
print("Last User ID:", last_user_id)

# Observations: Print the type() of required elements
print(type(api_response["data"]))
print(type(api_response["data"][0]["info"]))
```

```python
#Task 3: String Sanitization & Cleaning Pipeline

# Raw data
raw_log = " ERROR CODE: 404 | STATUS: NOT_FOUND | SOURCE: SERVER_01 "

# Pipeline step 1: Remove leading and trailing whitespace
clean_log = raw_log.strip()

# Pipeline step 2: Replace all underscores (_) with spaces
clean_log = clean_log.replace("_", " ")

# Locate the index of the vertical bar |
pipe_index = clean_log.find("|")
print("Index of first '|':", pipe_index)

# Convert the entire string to lowercase
clean_log = clean_log.lower()

# Verification
print("Cleaned Log:", clean_log)
```

```python

#Task 4: The CSV-to-Python Parser (Split & Join)

# CSV data
csv_data = "Apple, iPhone, 1200, Silver"

# Split into list
product_details = csv_data.split(", ")
print(product_details)

# Change the price (index 2) to "1300"
product_details[2] = "1300"
print(product_details)

# Reconstruction: Join the list using a semicolon (;)
reconstructed_string = ";".join(product_details)
print(reconstructed_string)

# Create header and concatenate using newline (\n)
header = "BRAND; MODEL; PRICE; COLOR"
final_output = header + "\n" + reconstructed_string
print(final_output)
```

```python
#Task 5:Atomic Dictionary Merging & Popping

personal_info = {"name": "John", "age": 30}
job_info = {"salary": 5000, "role": "Analyst"}

# Merging Dictionaries (Union Operator |)
full_profile = personal_info | job_info
print("Full Profile:", full_profile)

# Extraction Using .pop()
removed_age = full_profile.pop("age")
print("Removed Age:", removed_age)
print("Profile after pop:", full_profile)

# Clearing a Dictionary Using .clear()
personal_info.clear()
print("Length of personal_info:", len(personal_info))
```
