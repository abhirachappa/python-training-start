```python
# Task 1: Fundamental Conditional Logic (if-else)

def get_movie_ticket_type(age):
    """
    Determines ticket type based on age.
    If 18 or older: Regular Ticket
    If under 18: Child Discount Ticket
    """
    if age >= 18:
        return "Eligible for Regular Ticket"
    else:
        return "Eligible for Child Discount Ticket"

print(f"Age 20: {get_movie_ticket_type(20)}")
print(f"Age 15: {get_movie_ticket_type(15)}")
print(f"Age 18: {get_movie_ticket_type(18)}")
print(f"Age 5:  {get_movie_ticket_type(5)}")
```

```python
# Task 2: Multi-Branch Decision Making (if-elif-else)

def calculate_shipping_cost(order_amount):
    if order_amount >= 100.00:
        return 0.0
    
    elif order_amount >= 50.00:
        return 5.0
    
    else:
        return 10.0
 
print(f"\nOrder $120.50: ${calculate_shipping_cost(120.50)}")
print(f"Order $75.00:  ${calculate_shipping_cost(75.00)}")
print(f"Order $30.00:  ${calculate_shipping_cost(30.00)}")
print(f"Order $99.99:  ${calculate_shipping_cost(99.99)}")
print(f"Order $50.00:  ${calculate_shipping_cost(50.00)}")
```

```python
# Task 3: Iterating with for Loops and Sequences
# Problem Statement: "Character Frequency Counter"

def count_characters(text):
    char_counts = {}
    for char in text:
        if char in char_counts:
            char_counts[char] = char_counts[char] + 1
        else:
            char_counts[char] = 1
    return char_counts

print(f"Programming: {count_characters('Programming')}")
print(f"Python:      {count_characters('Python')}")
print(f"hello world: {count_characters('hello world')}")
```

```python
# Task 4: Dynamic Iteration with while Loops
# Problem Statement: "User Input Validator"

def get_positive_integer_input():
    simulated_inputs = ["-5", "0", "10"]
    input_index = 0
    
    while True:
        user_input = int(simulated_inputs[input_index])
        print(f"System Prompt: Enter a positive integer: {user_input}")
        if user_input > 0:
            return user_input 
        else:
            print("System Output: Invalid input. Please enter a positive integer.")
            
        input_index = input_index + 1

valid_number = get_positive_integer_input()
print(f"Final Result: Function returned {valid_number}")
```

```python
# Task 5: Complex Control Flow (Nested Conditionals & Loop Control)
# Problem Statement: "Prime Number Finder in a Range"

def find_primes_in_range(start, end):
    primes = []
    
    for num in range(start, end + 1):
        if num > 1:
            is_prime = True
            for i in range(2, num):
                if (num % i) == 0:
                    is_prime = False
                    break  
            if is_prime:
                primes.append(num)         
    return primes

print(f"Range 1-10:  {find_primes_in_range(1, 10)}")
print(f"Range 10-20: {find_primes_in_range(10, 20)}")
print(f"Range 1-1:   {find_primes_in_range(1, 1)}")
print(f"Range 2-2:   {find_primes_in_range(2, 2)}")
print(f"Range 20-30: {find_primes_in_range(20, 30)}")
```
