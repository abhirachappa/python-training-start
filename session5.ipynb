```python
# Task 3: Nested Conditionals & Iteration - For Loops

#Product Discount Eligibility

product_price = 120.0
customer_type = "premium"
print(f"Price: {product_price}, Customer: {customer_type}")
if customer_type == "premium":
    if product_price > 100:
        discount = 0.20  # 20% discount 
    else:
        discount = 0.10  # 10% discount 
elif customer_type == "regular":
    if product_price > 50:
        discount = 0.05  # 5% discount 
    else:
        discount = 0.0   # No discount 
else:
    discount = 0.0

final_price = product_price - (product_price * discount)
print(f"Final Price after discount: ${final_price}")

#Count Down Timer

print("\nStarting Countdown:")
for i in range(10, 0, -1):
    print(i)
print("Blast Off!")

#Sum of List Elements
numbers = [10, 25, 13, 40, 5, 30]
total_sum = 0
for n in numbers:
    total_sum = total_sum + n

print(f"\nNumbers List: {numbers}")
print(f"Total Sum: {total_sum}")
```

```python
# Task 4: Iteration - While Loops & Loop Control

#Password Attempt Limit 

correct_password = "secret"
attempts = 0
max_attempts = 3
test_passwords = ["123", "password", "secret"] 

print("Login System Simulation:")
while attempts < max_attempts:
    current_guess = test_passwords[attempts]
    print(f"Attempt {attempts + 1}: Entered '{current_guess}'")
    
    if current_guess == correct_password:
        print("Login successful!")
        break  
    attempts = attempts + 1

if attempts == max_attempts and current_guess != correct_password:
    print("Too many failed attempts. Account locked.")

#Collecting Positive Numbers

print("\nCollecting Positive Numbers:")
input_numbers = [10, 20, 5, -1] 
running_total = 0
index = 0

while index < len(input_numbers):
    num = input_numbers[index]
    print(f"Checking: {num}")
    if num > 0:
        running_total = running_total + num
    else:
        print("Non-positive number found. Stopping.")
        break 
    index = index + 1

print(f"Total of positive numbers: {running_total}")

#Filtering Even Numbers
print("\nFiltering Even Numbers (1-10):")
for x in range(1, 11):
    if x % 2 != 0:
        continue 
    print(x)
```

```python
# Task 5: Combined Concepts - Building a Simple Menu System

# Interactive Menu 

menu_choices = ["1", "2", "3"]
index = 0

print("--- Main Menu ---")
print("1. Perform Operation A")
print("2. Perform Operation B")
print("3. Exit")
print("-----------------")

while True:
    choice = menu_choices[index]
    print(f"Enter your choice: {choice}")
    
    if choice == "1":
        print("Performing Operation A...")
    elif choice == "2":
        print("Performing Operation B...")
    elif choice == "3":
        print("Exiting application. Goodbye!")
        break  # This stops the menu loop
    else:
        print("Invalid choice. Please try again.")
    
    index = index + 1

#Number Guessing Game (Basic)
secret_number = 7

guesses = [5, 10, 7]
g_index = 0

print("\n--- Number Guessing Game ---")
while True:
    guess = guesses[g_index]
    print(f"Guess the number: {guess}")
    
    if guess < secret_number:
        print("Too low!")
    elif guess > secret_number:
        print("Too high!")
    else:
        print("Congratulations! You guessed it!")
        break  
        
    g_index = g_index + 1

#Print Multiples of 3 (excluding 9)

print("\nMultiples of 3 (excluding 9) from 1 to 20:")
for num in range(1, 21):
    
    if num % 3 != 0:
        continue
        
    if num == 9:
        continue 
        
    print(num)
```
